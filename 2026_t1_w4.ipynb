# Variable Names
Technically, you can name your variable ALMOST anything you want. But good variable names describe what it contains. Imagine you went to a house and labelled all the boxes as 'Stuff'. You'd never be able to find anything.

Now, the reason I mentioned 'ALMOST' anything is because Python does have a few naming restrictions. You can name a variable anything as long as it follows three rules:

- It can only be one word with no spaces
- It can use only letters, numbers and the underscore (_) character
- It can't begin with a number

In the following program, incorrect variable names have rendered the program broken. Fix the issue with the variable names so that the code works!

```python
student name = "Alex"
2nd_score = 87
class-score = 92

print("Student:", student name)
print("Second Score:", 2nd_score)
print("Class Score:", class-score)
```

# String Concatenation
Although strings aren't numbers, you can still apply operators (e.g. +, -, x, /) to them! Operators change depending on th edata type. 

For example + is the addition operator for integers or floats:

```python
sum = int(3) + float(3.502)
print(sum)
print(type(sum))
```

```python
sum = 6 + 7
print(sum)
print(type(sum))
```

But when the '+' operator is used on two string values, it joins the strings as what is called the 'string concatenation' operator. For example:

```python
string_one = "Hello"
string_two = "there"
print(string_one + string_two)
```

# Turtle Graphics
The turtle module lets Python control a small “turtle” that moves around the screen drawing lines. By giving the turtle instructions such as move forward, turn, and change colour, we can create pictures using code. NOTE: DO NOT RUN TURTLE IN THE PYTHON NOTEBOOK, Take it to a .py file instead!

```python
import turtle

t = turtle.Turtle()

# t is now our turtle object. We can give it commands
t.forward(100)
t.left(90)
t.forward(100)

turtle.done()
```

## Loops: Repeating Instructions
Often we want the turtle to repeat actions multiple times. Instead of writing the same code again and again, we can use a loop. The following program draws a square by repeating 4 times.

```python
import turtle

t = turtle.Turtle()

for i in range(4):
    t.forward(100)
    t.left(90)

turtle.done()
```

## Turtle colors
The turtle can draw using different colours depending on what you set its color variable to.

```python
import turtle

t = turtle.Turtle()

t.color("red")
t.forward(100)

t.color("blue")
t.left(90)
t.forward(100)

turtle.done()
```

## Challenge: 
Make a rainbow spiral.

Hints:
- Use a list of colours
- Change the colour inside the loop
- Slowly increase the distance the turtle moves.

```python
import turtle

t = turtle.Turtle()

colours = ["red","orange","yellow","green","blue","purple"]

for i in range(60):
    t.color(colours[i % 6])
    t.forward(i * 5)
    t.left(59)

turtle.done()
```
